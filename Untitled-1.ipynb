```python
class Robot:
    def __init__(self):
        self.sentence = "안뇽"
        #self는 만들어진 인스턴스 자기자신을 자체를 가리킴
        #self. 무엇은 self 내부의 것(변수, 함수)를 가르킨다
        #클래스 내부의 변수는 속성(attribute)라고 한다

    def greeting(self):
        print(self.sentence) #주머니에서 sentence를 꺼내 쓴다
        #print(sentence) #error

bot1 = Robot()
bot1.greeting()


        
```

```python
class Robot:
    def __init__(self, height, sentence): #밖에서 넘어오는 매개변수는 무조건 self 뒤에
        self.height = height #self.height는 밖에서 들어온 height랑은 다르다
        # 클래스 내부에서 써먹으려고 주머니에 넣은 것
        self.sentence = sentence

    def greeting(self):
        print(self.sentence)
        self.height -= 10
        print(f"제 키는 {self.height}이에요")
        

bot1 = Robot(160, "안녕")
bot1.greeting()
```

```python
#메서드 오버라이딩(overwriting)
class Robot:
    def greeting(self):
        print("안녕하세요")

class Study_bot(Robot):
    def greeting(self):
        super().greeting() #super()가 부모 클래스를 가리킴 #부모 것을 이렇게 사용하는 건 가능하다
        
        print("공부하는 로봇") #부모의 원래 기능은 유지하고 거기에 덧붙이고 싶을 때 오버라이딩 사용 가능

bot = Study_bot()
bot.greeting()
```

```python
class Robot:
    def func1(self):
        print("부모 클래스 func1 실행")
        self.money = 100

class Study_bot(Robot):
    def func2(self):
        print("부모 클래스 func2 실행")
        self.book = 10
    def func3(self):
        print("0운")
        self.book = 20

bot1 = Study_bot()
bot1.func3()
bot1.func1()
print(bot1.book)
print(bot1.money)
```

```python
class Robot:
    def __init__(self):
        print("부모 클래스의 init 실행")
        self.money=100


class study_robot:
    def __init__(self):
        print("자식 클래스 init 실행")
        self.book = 10

bot1 = study_robot()
print(bot1.book)

bot1.__init__()
print(bot1.money)
```

```python
class Robot:
    def greeting(self):
        print("로봇입니다")

class Life:
    def eat(self):
        print("냠냠")

class Study_bot(Robot, Life):
    def study(self):
        print("열심열심")


bot1 = Study_bot()
bot1.greeting()
bot1.eat()
bot1.study()
```
